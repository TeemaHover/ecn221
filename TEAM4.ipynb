```python
import pandas as pd

df = pd.read_csv("Bus.csv")

df
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
numeric_abalone = df.select_dtypes(include=['float64', 'int64'])
for feature in df.columns: 
    if df[feature].dtype == 'object': 
        df[feature] = pd.Categorical(df[feature]).codes
correlation_matrix = df.corr()
plt.figure(figsize=(10, 8))
sns.heatmap(numeric_abalone.corr(), cmap='RdYlGn', center=0, annot=True)
plt.title('Correlation Matrix of Numeric Features in Bus Dataset')
plt.show()
correlation_matrix
```
